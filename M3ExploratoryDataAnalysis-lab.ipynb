Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df['ConvertedComp'].describe(percentiles=[0.25, 0.5, 0.75])
```

```python
df['ConvertedComp'].max()
```

Find out the upper and lower bounds.


```python
# your code goes here
np.nanquantile(df['ConvertedComp'], [0.0, 0.5, 1.0])
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here

Q3 = np.nanquantile(df['ConvertedComp'], 0.75)
Q1 = np.nanquantile(df['ConvertedComp'], 0.25)
print(Q1,Q3)
```

```python
IQR = Q3 - Q1

LR = Q1 - 1.5 * IQR
UR = Q3 + 1.5 * IQR

count = 0
for i in df['ConvertedComp']:
    if (i<LR or i>UR):
        count = count + 1

print('Number of outliers:',count)
```

```python
print(LR,UR)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
df_clean = [(df['ConvertedComp'] < LR) & (df['ConvertedComp'] > UR)]
```

```python
# your code goes here

df_clean = [(df['ConvertedComp'] < LR) | (df['ConvertedComp'] > UR)]

df_clean 
```
