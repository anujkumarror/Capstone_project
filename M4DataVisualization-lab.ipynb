## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.head(5)
```

```python
df.reset_index()
df.drop(columns="index", inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
df['ConvertedComp'].plot(kind='hist', figsize=(15, 10), bins=20)
plt.xlabel("ConvertedComp")
plt.ylabel("Frequency")
plt.title("histogram for the column `ConvertedComp`")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df['Age'].plot(kind='box', figsize=(15, 10))
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Boxplot for the column `Age`")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


```python
df['WorkWeekHrs'].plot.box()
```

Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(15, 10))
plt.xlabel("Age")
plt.ylabel("Number of working hours per week")
plt.title("Scatter plot for Age vs working hours/week")
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

import seaborn as sns
sns.set_theme(style="dark")


sns.relplot(x="CodeRevHrs", y="WorkWeekHrs", hue="Age", size="Age",
            sizes=(40, 400), alpha=.7, palette="Spectral",
            height=6, data=df)

plt.xlabel("Number of hours")
plt.ylabel("Number of working hours per week")
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
stack_set2=df[["Age","ConvertedComp"]]
stack_set2.set_index("Age",inplace=True)

stack_set2.groupby('Age').median().take((np.arange(45,61,1, dtype=int))-16).plot.bar(stacked=True)
plt.title('Median hours, by age')
```

```python
stack_set3=df[["Age","ConvertedComp"]]
stack_set3.set_index("Age",inplace=True)

stack_set2.groupby('Age').median('ConvertedComp').take((np.arange(25,31,1, dtype=int))-16).plot.line()
plt.title('Median hours, by age')
```
